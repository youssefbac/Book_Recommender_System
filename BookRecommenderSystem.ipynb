## Popularity Recommender

```python
#merge rating and books datasets
data=ratings.merge(books,on='ISBN')
data.head()
```

```python
#group book title by rating number
num_rating=data.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating
```

```python
#group book title by average rating 
avg_rating=data.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating
```

```python
#create popularity dataset based on avg_rating and num_rating
pop_df=num_rating.merge(avg_rating,on='Book-Title')
pop_df
```

```python
#filter popular books with rating superior to 250
pop_df=pop_df[pop_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False)
```

```python
#merge books and pop_df datasets
#drop duplicate columns
#keep only relevant columns
pop_df=pop_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

## Colaborative Filtering 

```python
num_rating.head()
```

```python
#filter users with book rating > 200
x=data.groupby('User-ID').count()['Book-Rating'] > 200
#list of User ID
rate_users=x[x].index
```

```python

filter_rating=data[data['User-ID'].isin(rate_users)]
```

```python
#get books with rating superior than 50
y=filter_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books= y[y].index
```

```python
#get the dataset that filter books that have book rating superior than 50
final_rate=filter_rating[filter_rating['Book-Title'].isin(famous_books)]
```

```python
#create table with user-ID as columns and book_title as rows index and Book-Rating as values
pivot_tab=final_rate.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
#replace nan values with 0
pivot_tab.fillna(0,inplace=True)
```

```python
pivot_tab.head()
```

```python
#calculate similarity using user Book_Rating according to Book_Title
similarity= cosine_similarity(pivot_tab)
```

```python
similarity.shape
```

```python
def recommend(book_name):
    #get the index of the book_name
    index= np.where(pivot_tab.index==book_name)[0][0]
    #get the top 6 similar items to the first Book_Title
    similarity_items=sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similarity_items:
        item=[]
        temp_df= books[books['Book-Title']==pivot_tab.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))              
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('The Da Vinci Code')
```
